```python
# NumPy and Pandas Demo for Graduate Economics Study

# NumPy: Numerical Python
# Description: NumPy is a library for efficient numerical computations with arrays and matrices.
# Core Features: Fast array operations, linear algebra, random number generation.
# Applications: Simulation, econometric modeling, optimization, time series analysis.

import numpy as np

# Create a NumPy array
arr = np.array([1, 2, 3, 4, 5])
# Compute the mean
mean_val = np.mean(arr)
# Linear algebra: matrix multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 2]])
product = np.dot(A, B)

print("NumPy array:", arr)
print("Mean value:", mean_val)
print("Matrix product:\n", product)

# Pandas: Python Data Analysis Library
# Description: Pandas provides data structures for labeled, tabular data.
# Core Features: DataFrames, data cleaning, grouping, merging, time series.
# Applications: Data wrangling, panel data analysis, regression, empirical research.

import pandas as pd

# Create a DataFrame
data = {'GDP': [2.1, 2.3, 2.5], 'Inflation': [1.5, 1.7, 1.6]}
df = pd.DataFrame(data, index=['2018', '2019', '2020'])
# Calculate summary statistics
summary = df.describe()
# Select data for a specific year
gdp_2019 = df.loc['2019', 'GDP']

print("\nPandas DataFrame:\n", df)
print("\nSummary statistics:\n", summary)
print("\nGDP in 2019:", gdp_2019)

# Notes:
# - NumPy is ideal for mathematical operations and simulations.
# - Pandas excels at handling real-world datasets, cleaning, and analysis.
# - Both are essential for empirical economics research and data-driven modeling.
```
