```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV, LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline

# ISLP tools
from ISLP import load_data
from ISLP.models import ModelSpec as MS

# Load Auto dataset
url = "https://raw.githubusercontent.com/fedscornell/MachineLearning25/main/Auto.csv"
Auto = pd.read_csv(url, na_values='?').dropna()
Auto.head()
```

 ## Chapter 4 Question 14
### (a) Create binary variable `mpg01`
We create a new variable that equals 1 if `mpg` is above its median, and 0 otherwise.


```python
mpg_median = Auto['mpg'].median()
Auto['mpg01'] = (Auto['mpg'] > mpg_median).astype(int)
Auto.head()
```

```python
# (b)
sns.pairplot(Auto[['mpg01','mpg','cylinders','displacement','horsepower','weight','acceleration','year']], hue='mpg01')
plt.show()

cols = ['cylinders','displacement','horsepower','weight','acceleration','year']
for col in cols:
    sns.boxplot(x='mpg01', y=col, data=Auto)
    plt.title(f'{col} vs mpg01')
    plt.show()
```

**Findings:**  
The variables `cylinders`, `displacement`, `horsepower`, and `weight` are strongly negatively associated with `mpg01`.  
Cars with fewer cylinders, smaller displacement, lower horsepower, and lower weight tend to have higher gas mileage (`mpg01 = 1`).  
`year` shows a positive association  newer cars tend to be more fuel efficient.


```python
#(c)
predictors = ['cylinders','displacement','horsepower','weight','year']
X = Auto[predictors]
y = Auto['mpg01']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"Train size: {len(X_train)}, Test size: {len(X_test)}")
```

```python
#(d)
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
y_pred = lda.predict(X_test)

acc = accuracy_score(y_test, y_pred)
test_error = 1 - acc

print("LDA Test Accuracy:", acc)
print("LDA Test Error:", test_error)
```

**Interpretation:**  
The LDA classifier correctly predicts whether a car has high or low mileage about 87% of the time,  
with a test error rate of approximately 12.7%.


### (e) Quadratic Discriminant Analysis (QDA)
We now fit a QDA model using the same predictors (`cylinders`, `displacement`, `horsepower`, `weight`, `year`)  
that appeared most associated with `mpg01` in part (b). We report the test error rate.


```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

qda = QuadraticDiscriminantAnalysis()
qda.fit(X_train, y_train)
y_pred_qda = qda.predict(X_test)

acc_qda = accuracy_score(y_test, y_pred_qda)
test_error_qda = 1 - acc_qda

print("QDA Test Accuracy:", acc_qda)
print("QDA Test Error:", test_error_qda)
```

**Interpretation:**  
The QDA model typically achieves an accuracy around 0.88 (test error  0.12), which is quite close to that of the LDA model. Since QDA allows quadratic boundaries, it can capture slightly more complex patterns if they exist.


### (f) Logistic Regression
We fit a logistic regression model to predict `mpg01` using the same predictors.  
The response is binary, so we use the logistic model and compute the test error.


```python
from sklearn.linear_model import LogisticRegression

logit = LogisticRegression(max_iter=2000)
logit.fit(X_train, y_train)
y_pred_logit = logit.predict(X_test)

acc_logit = accuracy_score(y_test, y_pred_logit)
test_error_logit = 1 - acc_logit

print("Logistic Regression Test Accuracy:", acc_logit)
print("Logistic Regression Test Error:", test_error_logit)
```

**Interpretation:**  
Logistic regression gives an accuracy around 0.88, similar to LDA and QDA.  
This suggests that the relationship between predictors and `mpg01` is largely linear, so all linear-based classifiers perform comparably well.


### (g) Naive Bayes Classification
We now apply a Gaussian Naive Bayes classifier to the same training and test data,  
assuming that each predictor is conditionally independent given the class.


```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)

acc_nb = accuracy_score(y_test, y_pred_nb)
test_error_nb = 1 - acc_nb

print("Naive Bayes Test Accuracy:", acc_nb)
print("Naive Bayes Test Error:", test_error_nb)
```

**Interpretation:**  
The Naive Bayes model yields a test error around 0.127.  
Although its independence assumption is strong, it still performs reasonably well on this dataset.


### (h) K-Nearest Neighbors (KNN)
We fit KNN classifiers using several values of \( K \) to find the one that minimizes test error.  
All predictors are standardized before fitting, since KNN is distance-based.


```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

test_errors = []
k_values = range(1, 21)

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    y_pred_knn = knn.predict(X_test_scaled)
    test_errors.append(1 - accuracy_score(y_test, y_pred_knn))

plt.plot(k_values, test_errors, marker='o')
plt.xlabel('K')
plt.ylabel('Test Error')
plt.title('KNN Test Error vs K')
plt.show()

best_k = k_values[np.argmin(test_errors)]
best_error = min(test_errors)
print("Best K:", best_k)
print("Minimum Test Error:", best_error)
```

**Interpretation:**
The KNN test error decreases sharply from K=1 to K=2, reaching its minimum at **K=2** (test error  **0.0847**).  
For larger K values, the error fluctuates and generally increases beyond K=7.  
Therefore, **K=2 provides the best performance** on this dataset, achieving a test accuracy of about **91.5%**,  
which is slightly better than all other methods (LDA, QDA, logistic regression, and Naive Bayes).  
This indicates that a small number of nearest neighbors captures the non-linear relationship in the predictors effectively.

## Chapter 6 Question 1

### Conceptual  Chapter 6, Question 1

**(a) Which of the three models with _k_ predictors has the smallest training RSS?**

**Best subset selection**.  
Reason: best subset evaluates **all** \(\binom{p}{k}\) models of size \(k\) and picks the one with the lowest training RSS.  
Forward/backward stepwise only search a restricted path, so their training RSS at size \(k\) can be **no smaller** (it can tie but not beat).

---

**(b) Which of the three models with _k_ predictors has the smallest test RSS?**

**No method is guaranteed to win.**  
Reason: test RSS depends on the true data-generating process and sampling noise.  
Any of the three may achieve the lowest test RSS for a given dataset; there is **no dominance guarantee**.  
(Practically, one would compare by validation/CV.)

---

**(c) True or False (with brief explanations):**

i. *The predictors in the \(k\)-variable model identified by forward stepwise are a subset of those in the \((k+1)\)-variable model identified by forward stepwise.*  
**True.** Forward stepwise adds one variable at a time and never removes; the selected sets are nested along the **forward** path.

ii. *The predictors in the \(k\)-variable model identified by backward stepwise are a subset of those in the \((k+1)\)-variable model identified by backward stepwise.*  
**True.** Backward stepwise starts from the full model and removes one at a time; thus the \(k\)-variable set is nested within the \((k+1)\)-variable set along the **backward** path.

iii. *The predictors in the \(k\)-variable model identified by backward stepwise are a subset of those in the \((k+1)\)-variable model identified by forward stepwise.*  
**False.** Forward and backward paths need not coincide; there is **no cross-method nesting** guarantee.

iv. *The predictors in the \(k\)-variable model identified by forward stepwise are a subset of those in the \((k+1)\)-variable model identified by backward stepwise.*  
**False.** Again, different search paths  no guaranteed nesting across methods.

v. *The predictors in the \(k\)-variable model identified by best subset selection are a subset of those in the \((k+1)\)-variable model identified by best subset selection.*  
**False.** Best subset chooses the **best model independently at each size**; the selected sets **need not be nested** (variables can be swapped as \(k\) changes).

### Chapter 6, Question 2

**(a) The lasso, relative to least squares**

 **Answer: (iii)**  

**Explanation:**  
The lasso is **less flexible** than least squares because it imposes an L1 penalty that shrinks some coefficients toward zero, reducing variance but increasing bias.  
It will therefore improve prediction accuracy when its **increase in bias** is less than its **decrease in variance**.

---

**(b) Ridge regression, relative to least squares**

 **Answer: (iii)**  

**Explanation:**  
Ridge regression is also **less flexible** than least squares because it adds an L2 penalty, shrinking coefficients but not setting them exactly to zero.  
Like the lasso, ridge improves prediction accuracy when the **reduction in variance** outweighs the **increase in bias**.

---

**(c) Non-linear methods, relative to least squares**

 **Answer: (i)**  

**Explanation:**  
Non-linear methods (e.g., polynomial regression, splines, or tree-based models) are **more flexible** than least squares, generally reducing bias but increasing variance.  
Such models will improve prediction accuracy when the **increase in variance** is smaller than the **decrease in bias**.
